10. (12б) Визначити функцію, яка приймає на вході масив і транспонує його. Скористайтесь методом reshape. І не забудьте почати з docstrings.

```python
def reshape_reverse(input_array):
    """
    Returns a reshaped version of the input array with reversed shape dimensions.

    Args:
        input_array (numpy.ndarray): The array to be reshaped.

    Returns:
        numpy.ndarray: A reshaped array with reversed shape dimensions.
    """
    reversed_shape = input_array.shape[::-1]
    reshaped_array = input_array.reshape(reversed_shape)
    return reshaped_array

print(reshape_reverse(colum_vector)) # pseudo-transpose
print(reshape_reverse(reshape_reverse(colum_vector))) # back to original

matrix_example = np.array([[1, 2, 3],
                           [4, 5, 6]])
print(reshape_reverse(matrix_example))
```

11. (7б) Створіть два масиви. Використовучи ці масиви, продемонструйте 1) операцію по-елементного додавання, 2) операцію по-елементного віднімання, 3) множення масиву на число, 4) операцію по-елементного множення, і 5) матричного множення.

```python
array_one = np.array([[1, 2],
                      [3, 4]])

array_two = np.array([[5, 6],
                      [7, 8]])

print("Sum: \n",array_one + array_two)
print("Diff: \n",array_one - array_two)
print("Multiply by 2: \n",array_one*2)
print("Multiply by elements: \n",array_one * array_two)
print("Multiply matrices: \n",np.dot(array_one, array_two))
```
